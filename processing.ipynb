```python
import pandas as pd
import numpy as np
```

```python
# Read in the data
df = pd.read_csv('mobiles.csv')
df.info()
df
```

```python
def clean_price_data(price):
    if isinstance(price, str):
        price = price.replace(',','').replace('₹','').strip()
    if price.isnumeric():
        return float(price)
```

```python
def split_rr_data(rr):
    if isinstance(rr, str):
        rr = rr.split('&')
        return rr[0], rr[1]
    else:
        return np.nan, np.nan
```

```python
df['cleaned_price']=df['price'].apply(clean_price_data)
```

```python
df['ratings'], df['reviews'] = zip(*df['rr'].apply(split_rr_data)) # Unpacking the tuple
```
